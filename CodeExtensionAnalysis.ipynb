```python
#syntatic analysis on code (if a generated code can be parsed into AST with no errors then valid)
#semantic passes already can be found in the metrics_all.jsonl for pass@1 (first guess correctness - run without crashing and first generated solution passes the unit test)

import json
import os
import ast

list_of_paths = [
    "BASEMODEL_CUSTOMROLECOLONCHATTEMP_ALL5DATASETS",
    "SFTMODEL_FULLROLECOLONRUN_CHECKPOINTFINAL_additionalconfigsfromtransformers4library",
    "DPOMODEL_FULLSFTROLECOLONRUN_DPOTINKERCHECKPOINT3000_mergedwithbasemodel"
]

list_of_datasets = [
    "mbpp",
    "codex_humanevalplus"
]

total = 0
correct = 0

current_dataset = ""
#extracts generated code from predictions.jsonl and attempts to parse into valid AST (if not valid print out error)
for filename in list_of_paths:
    for dataset in list_of_datasets:

        path = ""
        if(dataset == "mbpp"):
            path = f"./{filename}-eval-{dataset}/task-000-mbpp-predictions.jsonl"
            current_dataset = "mbpp"
        elif(dataset == "codex_humanevalplus"):
            path = f"./{filename}-eval-{dataset}/task-000-codex_humanevalplus-predictions.jsonl"
            current_dataset = "codex_humanevalplus"
        try:
            with open(path, "r") as f:
                for line in f:
                    #load a prediction (one generated function at a time) and apply parse to it, if it passes then valid AST parse otherwise catch error and print out the parsing issue
                    data = json.loads(line)
                    pred = data["model_output"][0]["model_answer"]
                    total+=1
                    try:
                      ast.parse(pred)
                      correct+=1
                    except (SyntaxError, TypeError) as e:
                      print(e)

        except FileNotFoundError:
            print(f"NOTPARTOFPRINTOUT: Skipping missing file: {path}")
        print(f"\n\n=========== For {filename} on dataset {current_dataset}: {correct}/{total} valid parses\n")
        total = 0
        correct = 0
```
